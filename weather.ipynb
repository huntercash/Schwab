## Data Cleanup

#### Things to analyze
- Split Begin Year Month
- create a single column for YYYY-MM-DD for Quandl API comparison for begin, and end

```python
dtdf = pd.DataFrame(xlsx)
dtdf.head()
```

```python
dtdf['BEGIN_YM_STR'] = dtdf['BEGIN_YEARMONTH'].astype(str)
dtdf['BEGIN_YEAR'] = df['BEGIN_YM_STR'].str[0:4]
dtdf['BEGIN_MONTH'] = df['BEGIN_YM_STR'].str[4:6]
dtdf['END_YM_STR'] = dtdf['END_YEARMONTH'].astype(str)
dtdf['END_YEAR'] = df['END_YM_STR'].str[0:4]
dtdf['END_MONTH'] = df['END_YM_STR'].str[4:6]
dtdf["START_DATE"] = dtdf["BEGIN_YEAR"].map(str) + "-" + dtdf["BEGIN_MONTH"].map(str) + "-" +\
dtdf["BEGIN_DAY"].map(str)
dtdf["END_DATE"] = dtdf["END_YEAR"].map(str) + "-" + dtdf["END_MONTH"].map(str) + "-" +\
dtdf["END_DAY"].map(str)
df.head()
```

```python
dtdf.DAMAGE_PROPERTY = (dtdf.DAMAGE_PROPERTY.replace(r'[KM]+$', '', regex=True).astype(float) * \
dtdf.DAMAGE_PROPERTY.str.extract(r'[\d\.]+([KM]+)', expand=False).fillna(0).replace(['K','M'], [10**3, 10**6]).astype(int))
```

```python
# Whoa a Billion..
dtdf.DAMAGE_CROPS = (dtdf.DAMAGE_CROPS.replace(r'[KMB]+$', '', regex=True).astype(float) * \
dtdf.DAMAGE_CROPS.str.extract(r'[\d\.]+([KMB]+)', expand=False).fillna(0).replace(['K','M','B'], [10**3, 10**6, 10**9]).astype(int))
```

```python
dtdf['DAMAGE_PROPERTY'] = dtdf['DAMAGE_PROPERTY'].fillna(0)
dtdf['DAMAGE_CROPS'] = dtdf['DAMAGE_CROPS'].fillna(0)
dtdf.head()
```

```python
dtdf.columns
```

```python
final = dtdf[['EVENT_ID','START_DATE', 'BEGIN_TIME', 'END_DATE', 'END_TIME','DAMAGE_PROPERTY', 'DAMAGE_CROPS','EVENT_TYPE','CZ_NAME', 'STATE', 'SOURCE'\
             ,'EPISODE_ID', 'INJURIES_DIRECT', 'INJURIES_INDIRECT','DEATHS_DIRECT','DEATHS_INDIRECT','MAGNITUDE','FLOOD_CAUSE'\
             ,'TOR_F_SCALE', 'BEGIN_LAT','BEGIN_LON','END_LAT','END_LON']]
final.head()
```

```python
# Complete the call to convert the date column
final['START_DATE'] =  pd.to_datetime(final['START_DATE'], format='%Y-%m-%d')
final['END_DATE'] =  pd.to_datetime(final['END_DATE'], format='%Y-%m-%d')
final
```
